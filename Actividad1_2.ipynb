#Filtros a aplicar

```python
# CVE_CLPV   (Valores entre 1000 a 2000)

df['CVE_CLPV'] = pd.to_numeric(df['CVE_CLPV'], errors='coerce') #errors='coerce' para manejar cualquier valor que no sea numérico convirtiéndolo en NaN

Filtro1 = df[(df['CVE_CLPV'] >= 1000) & (df['CVE_CLPV'] <= 2000)]

print(Filtro1 )
```

```python
#CVE_VEND    (TOdas las claves exepto la clave 5 y 4

# Filtra las claves distintas de '5' y '4'
Filtro2 = df[~((df['CVE_VEND'] == '5') | (df['CVE_VEND'] == '4'))]

print(Filtro2)
```

```python
#FECHA_ENT  ( 13/04/2022)
Filtro3 = df[df['FECHA_ENT'] == '2022-04-12 00:00:00']

print(Filtro3)
```

```python
# en la columna CAN_TOT imprimir aquellas que tienen Cantidades menores a 5951.7  o con estatus 'E'
Filtro4 = df[(df['CAN_TOT'] < 5951.7) | (df['STATUS'] == 'E')]

Filtro4
```

```python
#imprimir solo las columnas de CVE_DOC, FECHA_ENT,FECHA_VEN y CAN_TOT


Filtro5 = df[['CVE_DOC', 'FECHA_ENT', 'FECHA_VEN', 'CAN_TOT']]

Filtro5
```

```python
# Filas 7001 a la 7099

Filtro6 = df.iloc[7000:7100]

Filtro6
```
